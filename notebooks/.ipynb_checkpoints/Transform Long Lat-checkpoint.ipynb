```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
pd_2015 = gpd.read_file('Oss - Den Bosch 2015.geojson')
pd_2016 = gpd.read_file('Oss - Den Bosch 2016.geojson')
pd_2017 = gpd.read_file('Oss - Den Bosch 2017.geojson')
```

```python
series_2015 = gpd.GeoSeries(pd_2015['geometry'])
series_2016 = gpd.GeoSeries(pd_2016['geometry'])
series_2017 = gpd.GeoSeries(pd_2017['geometry'])

series_2015.crs = {'init': 'epsg:28992', 'no_defs': True}
series_2016.crs = {'init': 'epsg:28992', 'no_defs': True}
series_2017.crs = {'init': 'epsg:28992', 'no_defs': True}

series_2015 = series_2015.to_crs(epsg=4326)
series_2016 = series_2016.to_crs(epsg=4326)
series_2017 = series_2017.to_crs(epsg=4326)

pd_2015['geometry'] = series_2015
pd_2016['geometry'] = series_2016
pd_2017['geometry'] = series_2017
```

```python
with open('oss-denbosch-2015.geojson', 'w') as f:
    f.write(pd_2015.to_json())
```

```python
pd_2015.plot()
plt.show()
```
